## ğŸ§© Signal Extraction Tool (ADK)
This module interprets the user's scenario description and converts it into a normalized (0â€“1) signal vector used by all decision agents.


````python
class ExtractSignalsTool(AgentTool):
    func: ClassVar = None

    def run(self, context: ToolContext) -> dict:
        text = context.args.get("text", "")
        if not text:
            return {"error": "No text provided"}

        # Use Gemini 2.0 flash inside the tool
        model = Gemini(model_name="gemini-2.0-flash")

        prompt = f"""
        Extract structured numeric signals (0.0 to 1.0) from this context.
        Return ONLY valid JSON in this schema:

        {{
          "team_capacity": float,
          "team_stress": float,
          "reliability_risk": float,
          "revenue_impact": float,
          "operational_load": float,
          "strategic_importance": float,
          "urgency": float
        }}

        CONTEXT:
        {text}
        """

        response = model.run(prompt).text.strip()
        match = re.search(r"\{.*\}", response, re.DOTALL)
        if not match:
            return {"error": "Invalid JSON", "raw": response}

        return json.loads(match.group(0))
````

## ğŸ¤– Multi-Agent Decision Panel (ADK Agents)
Each agent represents a departmentâ€™s viewpoint and evaluates organizational decisions from that perspective.


### ğŸ§‘â€ğŸ’» Tech Lead Agent  
Evaluates engineering risk, feasibility, and complexity.
### ğŸ“¦ Product Manager Agent  
Evaluates customer value, roadmap alignment, and business impact.
### ğŸ›¡ Risk Agent  
Evaluates reliability, incident probability, and system fragility.
### ğŸ’° Finance Agent  
Evaluates revenue, cost, and ARR opportunity impact.
### ğŸ”§ Operations Agent  
Evaluates on-call load, operational burden, and infrastructure stability.
### ğŸ‘¨â€ğŸ”§ Engineering Manager Agent  
Evaluates team morale, burnout, capacity loss, and sustainability.


````python


# ---- Tech Lead ----
class TechLeadTool(AgentTool):
    func: ClassVar = None

    def run(self, context: ToolContext) -> dict:
        signals = context.args.get("signals", {})
        options = context.args.get("options", [])

        results = {}
        for opt in options:
            score = (
                (1 - signals["reliability_risk"]) * 0.4 +
                signals["team_capacity"] * 0.2 +
                (1 - signals["operational_load"]) * 0.2 +
                (1 - signals["team_stress"]) * 0.2
            )
            results[opt] = {
                "tech_lead_score": round(score, 3),
                "reasoning": f"Engineering prioritizes stability and team health for {opt}."
            }
        return results


# ---- Product Manager ----
class PMTool(AgentTool):
    func: ClassVar = None

    def run(self, context: ToolContext) -> dict:
        signals = context.args.get("signals", {})
        options = context.args.get("options", [])

        results = {}
        for opt in options:
            score = (
                signals["revenue_impact"] * 0.4 +
                signals["strategic_importance"] * 0.35 +
                signals["urgency"] * 0.25
            )
            results[opt] = {
                "pm_score": round(score, 3),
                "reasoning": f"PM emphasizes revenue, strategy, and urgency for {opt}."
            }
        return results


# ---- Risk Agent ----
class RiskTool(AgentTool):
    func: ClassVar = None

    def run(self, context: ToolContext) -> dict:
        signals = context.args.get("signals", {})
        options = context.args.get("options", [])

        results = {}
        for opt in options:
            score = (
                signals["reliability_risk"] * 0.5 +
                signals["team_stress"] * 0.3 +
                signals["operational_load"] * 0.2
            )
            results[opt] = {
                "risk_score": round(score, 3),
                "reasoning": f"Risk focuses on likelihood of incidents and fragility for {opt}."
            }
        return results


# ---- Finance Agent ----
class FinanceTool(AgentTool):
    func: ClassVar = None

    def run(self, context: ToolContext) -> dict:
        signals = context.args.get("signals", {})
        options = context.args.get("options", [])

        results = {}
        for opt in options:
            score = (
                signals["revenue_impact"] * 0.5 +
                signals["strategic_importance"] * 0.3 +
                (1 - signals["operational_load"]) * 0.2
            )
            results[opt] = {
                "finance_score": round(score, 3),
                "reasoning": f"Finance looks at revenue and long-term value for {opt}."
            }
        return results


# ---- Operations Agent ----
class OpsTool(AgentTool):
    func: ClassVar = None

    def run(self, context: ToolContext) -> dict:
        signals = context.args.get("signals", {})
        options = context.args.get("options", [])

        results = {}
        for opt in options:
            score = (
                (1 - signals["operational_load"]) * 0.4 +
                (1 - signals["reliability_risk"]) * 0.3 +
                signals["team_capacity"] * 0.2 +
                signals["urgency"] * 0.1
            )
            results[opt] = {
                "ops_score": round(score, 3),
                "reasoning": f"Ops focuses on load, reliability, and team capacity for {opt}."
            }
        return results


# ---- Engineering Manager ----
class EMTool(AgentTool):
    func: ClassVar = None

    def run(self, context: ToolContext) -> dict:
        signals = context.args.get("signals", {})
        options = context.args.get("options", [])

        results = {}
        for opt in options:
            score = (
                (1 - signals["team_stress"]) * 0.3 +
                signals["team_capacity"] * 0.25 +
                (1 - signals["reliability_risk"]) * 0.2 +
                signals["strategic_importance"] * 0.15 +
                (1 - signals["operational_load"]) * 0.1
            )
            results[opt] = {
                "em_score": round(score, 3),
                "reasoning": f"EM evaluates team health, long-term risk, and alignment for {opt}."
            }
        return results

print("âœ… All scoring tools loaded")
````

## ğŸ§© Supervisor Agent (Gemini 2.5 Flash/Pro)
The Supervisor reviews all agent reports, resolves conflicts, and produces the final decision, reasoning, and confidence score.


````python
class SupervisorAgent(LlmAgent):
    func: ClassVar = None

    def run(self, context):
        text = context.args.get("text", "")
        if not text:
            return {"error": "No context provided"}

        # Extract options automatically from text
        lines = text.splitlines()
        options = [l.strip() for l in lines if l.strip().lower().startswith("ship")]

        # Step 1 - Extract signals
        signals = self.call_tool("ExtractSignalsTool", args={"text": text})

        # Step 2 - Call all scoring tools
        panel = {
            "signals": signals,
            "tech_lead": self.call_tool("TechLeadTool", args={"signals": signals, "options": options}),
            "product_manager": self.call_tool("PMTool", args={"signals": signals, "options": options}),
            "risk": self.call_tool("RiskTool", args={"signals": signals, "options": options}),
            "finance": self.call_tool("FinanceTool", args={"signals": signals, "options": options}),
            "operations": self.call_tool("OpsTool", args={"signals": signals, "options": options}),
            "engineering_manager": self.call_tool("EMTool", args={"signals": signals, "options": options})
        }

        # Step 3 - Supervisor LLM decides
        schema = """
        {
          "final_choice": "string",
          "reasoning": "string",
          "risk_summary": "string",
          "agent_votes": {
            "tech_lead": "string",
            "product_manager": "string",
            "risk": "string",
            "finance": "string",
            "operations": "string",
            "engineering_manager": "string"
          },
          "confidence_score": float
        }
        """

        prompt = f"""
        You are the Supervisor Agent of a multi-agent decision system.

        You MUST return ONLY valid JSON. 
        Do NOT include reasoning outside the JSON. 
        Do NOT include markdown. 
        Do NOT include explanation before or after JSON. 
        Respond with JSON only.

        OPTIONS:
        {options}

        PANEL:
        {json.dumps(panel)}

        Return JSON ONLY in this exact format:

        {schema}
        """
````

````python
supervisor_agent = SupervisorAgent(
    name="supervisor_agent",
    model=Gemini(model_name="gemini-2.0-flash"),
    tools=[]
)


runner = InMemoryRunner(
    app_name="multi_agent_decision_app",
    agent=supervisor_agent
)

print("âœ… Supervisor and Runner ready")
````

````python
# FINAL ADK RUNNER â€“ Kaggle safe (no top-level await)

import asyncio

async def run_decision_async(context_text: str):
    input_text = context_text

    result = await runner.run_debug(input_text)

    # result is a list of ADK events â†’ last event contains the LLM text
    event = result[-1]
    raw_text = event.content.parts[0].text

    match = re.search(r"\{.*\}", raw_text, re.DOTALL)
    if not match:
        raise ValueError("Supervisor did not return JSON.")

    decision = json.loads(match.group(0))
    return decision, raw_text


print("ğŸš€ ADK Runner ready (Kaggle async-safe)")

````

````python
# === CLEAN MCKINSEY-STYLE FINAL DECISION FORMATTER ===

from IPython.display import Markdown, display
import re

def format_supervisor_output_mckinsey(raw_text):
    """
    Turns your supervisor's long explanation into a clean,
    executive-style report: Recommendation / Rationale / Next Steps.
    """
    # Normalize newlines
    text = raw_text.replace("\\n", "\n").strip()

    # Extract recommendation (first bold section)
    # Example pattern: **Negotiate Timeline Extension with Customer**
    m = re.search(r"\*\*(.*?)\*\*", text)
    recommendation = m.group(1).strip() if m else "Final Recommendation"

    # Split sections based on numbered or bullet patterns
    rationale_start = text.find("1.")
    next_steps_start = text.find("Action Plan")
    
    rationale_section = text[rationale_start:next_steps_start].strip() if rationale_start != -1 else ""
    next_steps_section = text[next_steps_start:].strip() if next_steps_start != -1 else ""

    # Clean formatting: restore bullets, remove extra * markers
    rationale_section = rationale_section.replace("**", "").replace("*", "-")
    next_steps_section = next_steps_section.replace("**", "").replace("*", "-")

    # Build clean markdown
    final_markdown = f"""
# ğŸ§  Recommendation
### **{recommendation}**

---

# ğŸ“Œ Rationale
{rationale_section}

---


"""
    display(Markdown(final_markdown))
````

## ğŸ§ª End-to-End Interactive Scenario Pipeline
This pipeline allows the user to:
1. Describe a business scenario  
2. Input organizational signals (0â€“10)  
3. Input agent scores  
4. Generate radar/heatmap  
5. Produce the final supervisor decision  


````python
# === FULL PIPELINE: USER INPUT â†’ RADAR â†’ HEATMAP â†’ ADK DECISION ===

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

async def run_full_pipeline():

    print("\n==============================")
    print(" ENTER SCENARIO DESCRIPTION ")
    print("==============================\n")
    scenario = input("Describe your scenario:\n")

    # ------------------------------
    # 1. SIGNAL INPUT (0â€“10 scale)
    # ------------------------------
    print("\n==============================")
    print(" ENTER SIGNAL VALUES (0â€“10) ")
    print("==============================\n")
    
    def get_signal(name):
        return float(input(f"{name} (0â€“10): "))/10.0

    signals = {
        "team_capacity": get_signal("Team capacity"),
        "team_stress": get_signal("Team stress"),
        "reliability_risk": get_signal("Reliability risk"),
        "revenue_impact": get_signal("Revenue impact"),
        "operational_load": get_signal("Operational load"),
        "strategic_importance": get_signal("Strategic importance"),
        "urgency": get_signal("Urgency"),
    }

    # ------------------------------
    # 2. AGENT SCORES INPUT (0â€“10)
    # ------------------------------
    print("\n==============================")
    print(" ENTER AGENT SCORES (0â€“10) ")
    print("==============================\n")

    options = ["Option A", "Option B"]
    agents = ["TechLead", "PM", "Risk", "Finance", "Ops", "EM"]

    panel = {agent: {} for agent in agents}

    for agent in agents:
        print(f"\n-- {agent} --")
        for opt in options:
            val = float(input(f"{agent} score for {opt} (0â€“10): "))
            panel[agent][opt] = val/10.0  # convert to 0â€“1

    df = pd.DataFrame(panel).T

    # ------------------------------
    # 3. VISUAL: RADAR CHART
    # ------------------------------
    labels = list(signals.keys())
    values = list(signals.values())

    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
    values += values[:1]
    angles += angles[:1]

    plt.figure(figsize=(6,6))
    ax = plt.subplot(111, polar=True)
    ax.plot(angles, values)
    ax.fill(angles, values, alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)
    plt.title("Signal Radar Chart")
    plt.show()

    # ------------------------------
    # 4. VISUAL: HEATMAP
    # ------------------------------
    plt.figure(figsize=(6,4))
    plt.imshow(df.values, aspect='auto')
    plt.xticks(ticks=np.arange(len(df.columns)), labels=df.columns)
    plt.yticks(ticks=np.arange(len(df.index)), labels=df.index)
    plt.colorbar()
    plt.title("Agent Score Heatmap")
    plt.show()

    print("\nAgent Score Table:")
    display(df)

    # # ------------------------------
    # # 5. ADK SUPERVISOR FINAL DECISION
    # # ------------------------------
    # print("\n==============================")
    # print(" ADK SUPERVISOR FINAL DECISION ")
    # print("==============================\n")

    result = await runner.run_debug(scenario)
    event = result[-1]
    supervisor_output = event.content.parts[0].text

    format_supervisor_output_mckinsey(supervisor_output)

    return 


# RUN IT
await run_full_pipeline()
````
